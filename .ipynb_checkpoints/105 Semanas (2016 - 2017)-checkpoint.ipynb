```python
melt2 = df.copy()
melt2['Last_Week_Sales'] = melt2.groupby(['PRODUTO'])['QUANT_VENDIDA'].shift()
melt2['Last_Week_Diff'] = melt2.groupby(['PRODUTO'])['Last_Week_Sales'].diff()
melt2 = melt2.dropna()
melt2.head(40)
```

```python
def rmsle(ytrue, ypred):
    return np.sqrt(mean_squared_log_error(ytrue, ypred))
    #return np.sqrt(mean_squared_error(ytrue, ypred))
```

```python
mean_error = []
for week in range(90,105):
    train = melt2[melt2['SEMANA'] < week]
    val = melt2[melt2['SEMANA'] == week]
    
    p = val['Last_Week_Sales'].values
    
    error = rmsle(val['QUANT_VENDIDA'].values, p)
    print('Week %d - Error %.5f' % (week, error))
    mean_error.append(error)
print('Mean Error = %.5f' % np.mean(mean_error))
```

```python
mean_error = []
for week in range(90,105):
    train = melt2[melt2['SEMANA'] < week]
    val = melt2[melt2['SEMANA'] == week]
    
    xtr, xts = train.drop(['QUANT_VENDIDA'], axis=1), val.drop(['QUANT_VENDIDA'], axis=1)
    ytr, yts = train['QUANT_VENDIDA'].values, val['QUANT_VENDIDA'].values
    
    mdl = RandomForestRegressor(n_estimators=1000, n_jobs=-1, random_state=0)
    mdl.fit(xtr, ytr)
    
    p = mdl.predict(xts)
    
    error = rmsle(yts, p)
    print('Week %d - Error %.5f' % (week, error))
    mean_error.append(error)
print('Mean Error = %.5f' % np.mean(mean_error))
```

```python
melt4 = df.copy()
melt4['Last_Week_Sales'] = melt4.groupby(['PRODUTO'])['QUANT_VENDIDA'].shift()
melt4['Last_Week_Diff'] = melt4.groupby(['PRODUTO'])['Last_Week_Sales'].diff()
melt4['Last-1_Week_Sales'] = melt4.groupby(['PRODUTO'])['QUANT_VENDIDA'].shift(2)
melt4['Last-1_Week_Diff'] = melt4.groupby(['PRODUTO'])['Last-1_Week_Sales'].diff()
melt4['Last-2_Week_Sales'] = melt4.groupby(['PRODUTO'])['QUANT_VENDIDA'].shift(3)
melt4['Last-2_Week_Diff'] = melt4.groupby(['PRODUTO'])['Last-2_Week_Sales'].diff()
melt4 = melt4.dropna()
melt4.head()
```

```python
mean_error = []
for week in range(90,105):
    train = melt4[melt4['SEMANA'] < week]
    val = melt4[melt4['SEMANA'] == week]
    
    xtr, xts = train.drop(['QUANT_VENDIDA'], axis=1), val.drop(['QUANT_VENDIDA'], axis=1)
    ytr, yts = train['QUANT_VENDIDA'].values, val['QUANT_VENDIDA'].values
    
    mdl = RandomForestRegressor(n_estimators=1000, n_jobs=-1, random_state=0)
    mdl.fit(xtr, np.log1p(ytr))
    
    p = np.expm1(mdl.predict(xts))
    
    error = rmsle(yts, p)
    print('Week %d - Error %.5f' % (week, error))
    mean_error.append(error)
print('Mean Error = %.5f' % np.mean(mean_error))
```
